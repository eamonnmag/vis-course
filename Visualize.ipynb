# Classic Data Interrogation using Pandas, Matplotlib, and Seaborn

```python
from pandas.tools.plotting import scatter_matrix
import pandas as pd
import numpy as np
%matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

**Data**

We're going to use a rather nice multi-variate dataset on categorisation of different species of Iris

```python
df = sns.load_dataset("iris")
```

## Getting an Overview of our data

### Scatter Plot Matrix with Kernel Density Estimate

```python
scatter_matrix(df, alpha=0.2, figsize=(6, 6), diagonal='kde')
```

### RadViz Matrix

Probably a bit confusing, but shows which dimensions in your data are driving a particular class separation

```python
from pandas.tools.plotting import radviz, parallel_coordinates, andrews_curves
radviz(df, 'Name',colormap='viridis')
```

### Parallel Coordinates

One of my favourites, but you should understand them now.

```python
parallel_coordinates(df, 'Name', colormap='viridis')
```

### Andrews Curves

Uses attributes in the data set as coefficients for the fourier series. 

```python
andrews_curves(df, 'Name')
```

### Correlations between variables in our dataset

```python
sns.heatmap(df.corr())
```

### Interrogating Distributions

```python
sns.distplot(df.PetalLength);
```

```python
sns.jointplot(x="PetalLength", y="PetalWidth", data=df);
```

```python
sns.jointplot(x="PetalLength", y="PetalWidth", data=df, kind="kde");
```

```python
sns.pairplot(df);
```

### Pairgrid



```python
g = sns.PairGrid(df)
g.map_diag(sns.kdeplot)
g.map_offdiag(sns.kdeplot, cmap="Blues_d", n_levels=6);
```
